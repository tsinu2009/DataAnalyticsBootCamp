```python
def score_to_numeric(x):
    if x=='I1':
        return 1
    if x=='SI2':
        return 2
    if x=='SI1':
        return 3
    if x=='VS2':
        return 4
    if x=='VS1':
        return 5
    if x=='VVS2':
        return 6
    if x=='VVS1':
        return 7
    if x=='IF':
        return 8

df['clarity_val'] = df['clarity'].apply(score_to_numeric)
df.tail()
```

```python
def score_to_numeric(x):
    if x=='Fair':
        return 1
    if x=='Good':
        return 2
    if x=='Very Good':
        return 3
    if x=='Premium':
        return 4
    if x=='Ideal':
        return 5
df['cut_val'] = df['cut'].apply(score_to_numeric)
df.tail()
```

```python
def score_to_numeric(x):
    if x=='J':
        return 1
    if x=='I':
        return 2
    if x=='H':
        return 3
    if x=='G':
        return 4
    if x=='F':
        return 5
    if x=='E':
        return 6
    if x=='D':
        return 7
    if x=='C':
        return 8
    if x=='B':
        return 9
    if x=='A':
        return 10
df['color_val'] = df['color'].apply(score_to_numeric)
df.tail()
```

```python
stats_df = df.drop (['cut','color','clarity'], axis=1)
stats_df.head()
```

```python
print('Estimated intercept coefficient:', lm.intercept_)
```
